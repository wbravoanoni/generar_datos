```python
import matplotlib.pyplot as plt
```

```python
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9 , 16, 25]
fig, ax = plt.subplots()
ax.plot(input_values,squares, linewidth=3)

# establecer el título de los gráficos y las etiquetas de los ejes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

#Establece el tamaño de las etieutas de los puntos de los ejes.
ax.tick_params(axis='both', which='major', labelsize=14)

#Ver estilos disponibles
"""plt.style.available"""

plt.style.use('seaborn-v0_8-paper')

#marcar puntos de interes , s stablecer el tamaño del punto
plt.scatter(2, 4, s=200)

plt.show()
```

# Serie de puntos con scatter()

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9,16, 25]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values)
```

```python
# Lo mismo pero con 100 puntos

x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()

"""
colores
c=(0, 0.8, 0
c='red'
"""
ax.scatter(x_values, y_values, s=10, c=y_values, cmap=plt.cm.Blues)
plt.style.use('seaborn-v0_8-paper')

#Establecer el rango para cada eje
ax.axis([0, 1100, 0, 1100000])

#Guardar imagen en la misma carpeta
#plt.savefig('squares_plot.png', bbox_inches='tight')
plt.show
```
